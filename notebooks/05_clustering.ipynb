```python
import pandas as pd
from sklearn.cluster import KMeans
import joblib

df = pd.read_csv("../data/final_data.csv")

store_data = df.groupby('Store_ID').agg({
    'Revenue':'sum',
    'Units_Sold':'sum',
    'Store_Rating':'mean'
})

kmeans = KMeans(n_clusters=3, random_state=42)
store_data['Cluster'] = kmeans.fit_predict(store_data)

joblib.dump(kmeans, "../models/kmeans_store_segmentation.pkl")

store_data.head()
```
